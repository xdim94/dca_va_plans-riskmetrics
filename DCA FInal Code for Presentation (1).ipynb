```python
import pandas as pd

def dca_portfolio_returns(df1, monthly_investment=100, max_investment=14400):
    """
    Calculates the returns for a monthly dollar-cost averaging investment strategy on a given dataframe.

    :param df: The dataframe containing the investment data.
    :param monthly_investment: The amount to invest each month. Default is 100.
    :param max_investment: The maximum amount to invest for each index. Default is 14400.
    :returns: A pandas dataframe with the monthly returns of the investment strategy.
    """
    # Convert the date column to a datetime object
    df1['Date'] = pd.to_datetime(df1['Date'])

    # Set the index column to be the date column
    df1.set_index('Date', inplace=True)

    # Filter the data to only include the first date of each month
    df1 = df1.resample('MS').first()

    # Set the initial investment amount and number of shares to zero for each index
    investment_amounts = {'EUSTX50': 0, 'CAC 40': 0, 'IBEX 35': 0, 'DAX': 0, 'IT40': 0}
    shares_owned = {'EUSTX50': 0, 'CAC 40': 0, 'IBEX 35': 0, 'DAX': 0, 'IT40': 0}

    # Create an empty dataframe to store the monthly returns
    returns_df1 = pd.DataFrame(index=df1.index, columns=investment_amounts.keys())

    # Loop through the data and invest the monthly amount on the first date of each month
    for index, row in df1.iterrows():
        for column in investment_amounts:
            # Calculate the number of shares to purchase with the monthly investment
            share_price = row[column]
            shares_purchased = monthly_investment / share_price

            # Add the shares to our total shares owned for this index
            shares_owned[column] += shares_purchased

            # Add the amount invested to our total investment amount for this index
            investment_amounts[column] += monthly_investment

            # Stop investing when the total investment amount for this index reaches the maximum investment amount
            if investment_amounts[column] > max_investment:
                break

        # Calculate the portfolio value and returns for each index for this month
        for column in investment_amounts:
            portfolio_value = shares_owned[column] * row[column]
            invested_amount = investment_amounts[column]
            returns = (portfolio_value - invested_amount) / invested_amount

            # Add the monthly returns to the returns dataframe
            returns_df1.loc[index, column] = returns

    return returns_df1


# Calculate the monthly DCA returns
returns_df1 = dca_portfolio_returns(df1)

# Print the monthly DCA returns
for index, row in returns_df1.iterrows():
    print(f"Monthly DCA returns for {index.strftime('%B %Y')}:")
    for column, value in row.items():
        print(f"{column}: {value:.2%}")
    print()
```

```python


def dca_portfolio_returns(df15, monthly_investment=100, max_investment=14400):
    """
    Calculates the returns for a monthly dollar-cost averaging investment strategy on a given dataframe.

    :param df: The dataframe containing the investment data.
    :param monthly_investment: The amount to invest each month. Default is 100.
    :param max_investment: The maximum amount to invest for each index. Default is 14400.
    :returns: A pandas dataframe with the monthly returns of the investment strategy.
    """
    # Convert the date column to a datetime object
    df15['A'] = pd.to_datetime(df15['A'])

    # Set the index column to be the date column
    df15.set_index('A', inplace=True)

    # Filter the data to only include the first date of each month
    df15 = df15.resample('MS').first()

    # Set the initial investment amount and number of shares to zero for each index
    investment_amounts = {'MSCI': 0}
    shares_owned = {'MSCI': 0}

    # Create an empty dataframe to store the monthly returns
    returns_df15 = pd.DataFrame(index=df15.index, columns=investment_amounts.keys())

    # Loop through the data and invest the monthly amount on the first date of each month
    for index, row in df15.iterrows():
        for column in investment_amounts:
            # Calculate the number of shares to purchase with the monthly investment
            share_price = float(row[column])

            shares_purchased = monthly_investment / share_price

            # Add the shares to our total shares owned for this index
            shares_owned[column] += shares_purchased

            # Add the amount invested to our total investment amount for this index
            investment_amounts[column] += monthly_investment

            # Stop investing when the total investment amount for this index reaches the maximum investment amount
            if investment_amounts[column] > max_investment:
                break

        # Calculate the portfolio value and returns for each index for this month
        for column in investment_amounts:
            portfolio_value = shares_owned[column] * float(row[column])
            invested_amount = investment_amounts[column]
            returns = (portfolio_value - invested_amount) / invested_amount

            # Add the monthly returns to the returns dataframe
            returns_df15.loc[index, column] = returns

    return returns_df15


# Calculate the monthly DCA returns
returns_df15 = dca_portfolio_returns(df15)

# Print the monthly DCA returns
for index, row in returns_df15.iterrows():
    print(f"Monthly DCA returns for {index.strftime('%B %Y')}:")
    for column, value in row.items():
        print(f"{column}: {value:.2%}")
    print()
```

```python
import pandas as pd
import statsmodels.api as sm

def calculate_jensen_alpha(merged_df99):
    # Calculate excess returns
    excess_returns = merged_df99.iloc[:, :-1].sub(merged_df99.iloc[:, -1], axis=0)
    
    # Calculate Jensen's alpha for each index
    alphas = []
    for i in range(5):
        X = sm.add_constant(excess_returns.iloc[:, -1].astype(float))
        y = excess_returns.iloc[:, i].astype(float)

        model = sm.OLS(y, X, missing='drop')
        results = model.fit()
        alphas.append(results.params[0])
    
    # Create a new dataframe to store the results
    alpha_df = pd.DataFrame({
        'Index': ['EUSTX50', 'CAC 40', 'IBEX 35', 'DAX', 'IT40'],
        'Jensen Alpha': alphas
    })
    
    return alpha_df
```

```python
import pandas as pd
import numpy as np

def calculate_sharpe_ratio(merged_df99):
    # Calculate excess returns
    excess_returns = merged_df99.iloc[:, :-2].sub(merged_df99.iloc[:, -2], axis=0)

    # Calculate annualized mean and standard deviation of excess returns
    mean_excess_returns = excess_returns.mean() * 12
    std_excess_returns = excess_returns.std() * np.sqrt(12)

    # Calculate Sharpe Ratio for each index
    sharpe_ratios = (mean_excess_returns / std_excess_returns)
    
    # Create a new dataframe to store the results
    sharpe_ratio_df = pd.DataFrame({
        'Index': ['EUSTX50', 'CAC 40', 'IBEX 35', 'DAX', 'IT40'],
        'Sharpe Ratio': sharpe_ratios.values
    })
    
    return sharpe_ratio_df
```

```python
def calculate_sortino_ratio(merged_df99, risk_free_rate):
    # Calculate excess returns
    excess_returns = merged_df99.iloc[:, :-1].sub(merged_df99.iloc[:, -1], axis=0)

    # Calculate downside deviation
    target_return = merged_df99.iloc[:, -1].mean()
    negative_excess_returns = excess_returns[excess_returns < target_return].stack()
    downside_deviation = np.sqrt(np.sum(negative_excess_returns.pow(2)) / negative_excess_returns.size)

    # Calculate annualized Sortino ratio for each index
    sortino_ratios = (merged_df99.iloc[:, :-2].mean() - risk_free_rate) / downside_deviation * np.sqrt(12)

    # Create a new dataframe to store the results
    sortino_ratio_df = pd.DataFrame({
        'Index': ['EUSTX50', 'CAC 40', 'IBEX 35', 'DAX', 'IT40'],
        'Sortino Ratio': sortino_ratios
    })

    return sortino_ratio_df
```

```python
import pandas as pd

def dca_portfolio_returns(merged_df4, monthly_investment=100, max_investment=14400):
    """
    Calculates the returns for a monthly dollar-cost averaging investment strategy on a given dataframe.

    :param df: The dataframe containing the investment data.
    :param monthly_investment: The amount to invest each month. Default is 100.
    :param max_investment: The maximum amount to invest for each index. Default is 14400.
    :returns: A pandas dataframe with the monthly returns of the investment strategy.
    """
    # Convert the date column to a datetime object
    merged_df4['Date'] = pd.to_datetime(merged_df4['Date'])

    # Set the index column to be the date column
    merged_df4.set_index('Date', inplace=True)

    # Filter the data to only include the first date of each month
    merged_df4 = merged_df4.resample('MS').first()

    # Set the initial investment amount and number of shares to zero for each index
    investment_amounts = {'SZSE GDP100': 0, 'JSE 25': 0, 'NIFTY50': 0, 'IMOEX': 0, 'IBOVESPA': 0}
    shares_owned = {'SZSE GDP100': 0, 'JSE 25': 0, 'NIFTY50': 0, 'IMOEX': 0, 'IBOVESPA': 0}

    # Create an empty dataframe to store the monthly returns
    returns_dfbrics = pd.DataFrame(index=merged_df4.index, columns=investment_amounts.keys())

    # Loop through the data and invest the monthly amount on the first date of each month
    for index, row in merged_df4.iterrows():
        for column in investment_amounts:
            # Calculate the number of shares to purchase with the monthly investment
            share_price = row[column]
            shares_purchased = monthly_investment / share_price

            # Add the shares to our total shares owned for this index
            shares_owned[column] += shares_purchased

            # Add the amount invested to our total investment amount for this index
            investment_amounts[column] += monthly_investment

            # Stop investing when the total investment amount for this index reaches the maximum investment amount
            if investment_amounts[column] > max_investment:
                break

        # Calculate the portfolio value and returns for each index for this month
        for column in investment_amounts:
            portfolio_value = shares_owned[column] * row[column]
            invested_amount = investment_amounts[column]
            returns = (portfolio_value - invested_amount) / invested_amount

            # Add the monthly returns to the returns dataframe
            returns_dfbrics.loc[index, column] = returns

    return returns_dfbrics


# Calculate the monthly DCA returns
returns_dfbrics = dca_portfolio_returns(merged_df4)

# Print the monthly DCA returns
for index, row in returns_dfbrics.iterrows():
    print(f"Monthly DCA returns for {index.strftime('%B %Y')}:")
    for column, value in row.items():
        print(f"{column}: {value:.2%}")
    print()
```

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np

def calculate_jensen_alpha2(merged_df102):
    # Select only the index returns columns
    index_returns = merged_df102.iloc[:, 1:-1]
       #adjusted_risk_free_rate = risk_free_rate2 / 12
    # Calculate Jensen's alpha and Sharpe ratio for each index
    alphasB = []
    sharpe_ratios = []
    for i in range(5):
        X = sm.add_constant(index_returns.iloc[:, -1].astype(float))
        y = index_returns.iloc[:, i].astype(float)

        model = sm.OLS(y, X, missing='drop')
        results = model.fit()
        alphasB.append(results.params[0])
        
        excess_returns = y - X.iloc[:, 1] * results.params[1]
        sharpe_ratio = np.mean(excess_returns) / np.std(excess_returns, ddof=1)
        sharpe_ratios.append(sharpe_ratio)
    
    # Create a new dataframe to store the results
    alpha_df2 = pd.DataFrame({
        'Index': ['SZSE GDP100', 'JSE 25', 'NIFTY50', 'IMOEX', 'IBOVESPA'],
        'Jensen Alpha': alphasB,
        'Sharpe Ratio': sharpe_ratios
    })
    
    return alpha_df2
```

```python
def calculate_sortino_ratio2(merged_df102, risk_free_rate2):
    # Calculate excess returns
    excess_returns2 = merged_df102.iloc[:, 1:-1].sub(merged_df102.iloc[:, -1], axis=0)

    # Calculate downside deviation
    target_return = merged_df102.iloc[:, -1].mean()
    negative_excess_returns = excess_returns2[excess_returns2 < target_return].stack()
    downside_deviation2 = np.sqrt(np.sum(negative_excess_returns.pow(2)) / negative_excess_returns.size)

    # Calculate annualized Sortino ratio for each index
    sortino_ratios2 = (merged_df102.iloc[:, 1:-2].mean() - risk_free_rate2) / downside_deviation2 * np.sqrt(12)

    # Create a new dataframe to store the results
    sortino_ratio_df2 = pd.DataFrame({
        'Index': ['SZSE GDP100', 'JSE 25', 'NIFTY50', 'IMOEX', 'IBOVESPA'],
        'Sortino Ratio': sortino_ratios2
    })

    return sortino_ratio_df2
```

```python
risk_free_rate2 = merged_df102['Cbond'].iloc[0]

sortino_ratio_df2 = calculate_sortino_ratio2(merged_df102, risk_free_rate2)
sortino_ratio_df2.index.name = 'Index'

print(sortino_ratio_df2)
```

```python
def calculate_treynor_ratio2(merged_df102, risk_free_rate2):
    # Select only the index returns columns
    index_returns22 = merged_df102.iloc[:, 1:-1]

    # Calculate Treynor ratio for each index
    treynor_ratios2 = []
    for i in range(5):
        excess_returns = index_returns22.iloc[:, i].astype(float) - risk_free_rate2
        beta = excess_returns.cov(merged_df102.iloc[:, -2].astype(float)) / merged_df102.iloc[:, -2].var()

        treynor_ratio = excess_returns.mean() / beta
        treynor_ratios2.append(treynor_ratio)

    # Create a new dataframe to store the results
    treynor_ratio_df22 = pd.DataFrame({
        'Index': ['SZSE GDP100', 'JSE 25', 'NIFTY50', 'IMOEX', 'IBOVESPA'],
        'Treynor Ratio': treynor_ratios2
    })

    return treynor_ratio_df22
```

```python
treynor_ratio_df22 = calculate_treynor_ratio2(merged_df102, risk_free_rate2)
```
