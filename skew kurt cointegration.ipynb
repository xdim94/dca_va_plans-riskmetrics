```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Assuming you have already calculated the monthly returns for all indices

# Create a figure with five subplots (1 row, 5 columns)
fig, axs = plt.subplots(1, 5, figsize=(25, 6))

# Plotting the histograms and normality curves for each index
indices = [DAX_monthly_returns, EUSTX50_monthly_returns, IBEX35_monthly_returns, IT40_monthly_returns, CAC40_monthly_returns]
index_names = ['DAX', 'EUSTX50', 'IBEX 35', 'FTSE MIB (IT40)', 'CAC 40']

for i in range(len(indices)):
    ax = axs[i]
    index_returns = indices[i]
    
    sns.histplot(index_returns.dropna(), kde=True, color='blue', bins=30, ax=ax)
    ax.set_title(f'Histogram of {index_names[i]} Monthly Returns')
    ax.set_xlabel('Returns')
    ax.set_ylabel('Frequency')

    # Plotting the normality curve
    mu, sigma = index_returns.mean(), index_returns.std()
    x = index_returns.dropna().values
    x = np.sort(x)
    p = norm.pdf(x, mu, sigma)
    ax.plot(x, p, 'k', linewidth=2, label='Normal Distribution')
    ax.legend()

    # Adding vertical lines for mean and median
    mean_val = index_returns.mean()
    median_val = index_returns.median()
    ax.axvline(mean_val, color='red', linestyle='dashed', linewidth=1.5, label='Mean')
    ax.axvline(median_val, color='green', linestyle='dashed', linewidth=1.5, label='Median')
    ax.legend()

plt.tight_layout()  # To ensure proper spacing between subplots

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, gaussian_kde

# Assuming you have already calculated the monthly returns for all indices

# Create a figure with five subplots (1 row, 5 columns)
fig, axs = plt.subplots(1, 5, figsize=(25, 6))

# Plotting the histograms and normality curves for each index
indices = [DAX_monthly_returns, EUSTX50_monthly_returns, IBEX35_monthly_returns, IT40_monthly_returns, CAC40_monthly_returns]
index_names = ['DAX', 'EUSTX50', 'IBEX 35', 'FTSE MIB (IT40)', 'CAC 40']

for i in range(len(indices)):
    ax = axs[i]
    index_returns = indices[i]
    
    sns.histplot(index_returns.dropna(), kde=True, color='blue', bins=30, ax=ax)
    ax.set_title(f'Histogram of {index_names[i]} Monthly Returns')
    ax.set_xlabel('Returns')
    ax.set_ylabel('Frequency')

    # Plotting the normality curve
    mu, sigma = index_returns.mean(), index_returns.std()
    x = index_returns.dropna().values
    x = np.sort(x)
    p = norm.pdf(x, mu, sigma)
    ax.plot(x, p, 'k', linewidth=2, label='Normal Distribution')
    ax.legend()

    # Adding vertical lines for mean, median, and mode (estimated from KDE)
    mean_val = index_returns.mean()
    median_val = index_returns.median()
    kde = gaussian_kde(index_returns.dropna())
    mode_val = x[np.argmax(kde(x))]  # Estimate mode from KDE
    ax.axvline(mean_val, color='red', linestyle='dashed', linewidth=1.5, label='Mean')
    ax.axvline(median_val, color='green', linestyle='dashed', linewidth=1.5, label='Median')
    ax.axvline(mode_val, color='purple', linestyle='dashed', linewidth=1.5, label='Mode (from KDE)')
    ax.legend()

plt.tight_layout()  # To ensure proper spacing between subplots

plt.show()
```

```python
import pandas as pd
from itertools import combinations
from statsmodels.tsa.stattools import adfuller

# Assume you have the monthly returns of each index in separate pandas Series.
# Replace the 'data_points' with the actual return values.


# Remove the first row (NaN) from each series
SZSE_monthly_returns = SZSE_monthly_returns.dropna()
JSE_monthly_returns = JSE_monthly_returns.dropna()
NIFTY50_monthly_returns = NIFTY50_monthly_returns.dropna()
IMOEX_monthly_returns = IMOEX_monthly_returns.dropna()
IBOVESPA_monthly_returns = IBOVESPA_monthly_returns.dropna()

# Combine all the returns into a dictionary
returns_dict = {
    'SZSE': SZSE_monthly_returns,
    'JSE': JSE_monthly_returns,
    'NIFTY50': NIFTY50_monthly_returns,
    'IMOEX': IMOEX_monthly_returns,
    'IBOVESPA': IBOVESPA_monthly_returns,
}

# List all possible combinations of returns for cointegration analysis
returns_combinations = list(combinations(returns_dict.keys(), 2))

# Perform ADF test for each pair of returns
for returns_1, returns_2 in returns_combinations:
    result = adfuller(returns_dict[returns_1] - returns_dict[returns_2])
    print(f'Difference between {returns_1} and {returns_2}')
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])
    print('Critical Values:', result[4])
    if result[1] <= 0.05:
        print(f"{returns_1} and {returns_2} are cointegrated.\n")
    else:
        print(f"{returns_1} and {returns_2} are not cointegrated.\n")
```
