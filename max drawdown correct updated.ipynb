```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in the data from index_brics_monthly_prices.csv file
df = pd.read_csv('index brics monthly prices.csv', index_col='Date', parse_dates=['Date'])

# Calculate drawdowns for each index
drawdowns = pd.DataFrame(index=df.index)
for column in df.columns:
    rolling_max = df[column].expanding().max()
    drawdowns[column] = (df[column] - rolling_max) / rolling_max

# Calculate maximum drawdown for each index
max_drawdowns = drawdowns.min()

# Convert drawdowns to percentage format
max_drawdowns_percent = max_drawdowns * 100

# Print the maximum drawdown for each index in percentage
print("Maximum Drawdowns (in percentage):")
print(max_drawdowns_percent)

# Calculate drawdown amount in euros (assuming initial investment of 14400 euros)
initial_investment = 14400
drawdown_amount_euros = (1 + max_drawdowns) * initial_investment - initial_investment

# Print the drawdown amount in euros for each index
print("\nDrawdown Amount (in euros):")
print(drawdown_amount_euros)

# Plot drawdowns for all 5 indices
plt.figure(figsize=(12, 6))
for column in df.columns:
    plt.plot(drawdowns.index, drawdowns[column], label=column)

plt.xlabel("Date")
plt.ylabel("Drawdown")
plt.title("Drawdowns for BRICS Indices")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in the data from europe_indices_monthly.csv file
df = pd.read_csv('europe indices monthly.csv', index_col='Date', parse_dates=['Date'])

# Calculate drawdowns for each index
drawdowns = pd.DataFrame(index=df.index)
for column in df.columns:
    rolling_max = df[column].expanding().max()
    drawdowns[column] = (df[column] - rolling_max) / rolling_max

# Calculate maximum drawdown for each index
max_drawdowns = drawdowns.min()

# Calculate annualized return (AR) for each index
months_in_data = len(df)
initial_value = df.iloc[0]
final_value = df.iloc[-1]
annualized_return = (final_value / initial_value) ** (12 / months_in_data) - 1

# Calculate Calmar ratio for each index
calmar_ratio = annualized_return / abs(max_drawdowns)

# Print the Calmar ratio for each index
print("Calmar Ratios:")
print(calmar_ratio)

# Plot drawdowns for all 5 indices
plt.figure(figsize=(12, 6))
for column in df.columns:
    plt.plot(drawdowns.index, drawdowns[column], label=column)

plt.xlabel("Date")
plt.ylabel("Drawdown")
plt.title("Drawdowns for EMU Indices")
plt.legend(loc="upper left")
plt.grid(True)
plt.show()
```
